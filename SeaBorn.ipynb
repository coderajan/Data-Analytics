# Seaborn Vs Matplotlib
--- Matplotlib doesn't server well when it comes to dealing with DataFrame while 
seaborn functions actually work on DataFrames 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
a=sns.load_dataset("flights")
sns.relplot(x='passengers',y='month',data=a)
# two data set 
```

```python
a=sns.load_dataset("flights")
sns.relplot(x='passengers',y='month',hue='year',data=a)
```

```python
b=sns.load_dataset('tips')
sns.relplot(x='time',y='tip',data=b)
```

```python
b=sns.load_dataset('tips')
sns.relplot(x='time',y='tip',data=b,kind='line')
```

# catPlot

```python
b=sns.load_dataset("tips")
sns.catplot(x='day',y='total_bill',data=b)
```

```python
b=sns.load_dataset("tips")
sns.catplot(x='day',y='total_bill',data=b,kind='box')
# WE can change the format of plot using the kind -- and pass value to kind 
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='violin')
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind="boxen")
```
